```python
import pandas as pd
```

```python
data = pd.read_csv(r'C:\Users\lenovo\Downloads\pokemon_data.csv')
data
```

```python
data['Total'] = data.iloc[:,4:10].sum(axis=1)
data
```

```python
cols = list(data.columns)
```

```python
data_new = data[cols[0:10] + [cols[-1]] + cols[10:12]]
data_new
```

### Save the data

```python
# saving as csv file
data_new.to_csv('Modified_Pokemon_data.csv',index=False)
```

```python
# saving as excel file
data_new.to_excel('Modified_Pokemon_data.xlsx',index=False)
```

```python
# saving as text file
data_new.to_csv('Modified_Pokemon_data.txt', index=False, sep='\t')
```

### Filtering Data

```python
df = data_new.copy()
df
```

```python
df.loc[df['Type 1'] == 'Grass']
```

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

```python
df_filtered = df.loc[((df['Type 1'] == 'Grass') | (df['Type 1'] == 'Fire')) & (df['Attack'] > 50)\
                    & (df['Defense'] > 50)]
df_filtered
```

```python
# Reset Index

dff = df_filtered.reset_index()
dff
```

```python
dff = df_filtered.reset_index(drop=True)
dff
```

```python
df_filtered.reset_index(drop=True,inplace=True)
df_filtered
```

```python
df.loc[df['Name'].str.contains('Mega')]
```

```python
df.loc[~df['Name'].str.contains('Mega')]
```

### Conditional Changes

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1' ] = 'Flamer'
df
```

```python
df.loc[df['Type 1'] == 'Flamer', 'Legendary'] = True
df
```

```python
df.loc[df['Total'] > 500, ['Generation','Legendary']] = ['Test 1','Test 2']
df
```
